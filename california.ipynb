```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
california_data = fetch_california_housing()
```

```python
california_data
```

```python
print(california_data.feature_names)
```

```python
dataset = pd.DataFrame(california_data.data, columns=california_data.feature_names)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['price'] = california_data.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()  # no null value
```

```python
dataset.describe()
```

```python
x = dataset.drop('price', axis=1)
y = dataset['price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model.fit(x_train, y_train)
```

```python
regression_model.score(x_train, y_train)
```

```python
# y = ax1 + bx2 + cx3 +........+ hx8 + i
```

```python
regression_model.coef_
```

```python
regression_model.intercept_
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
regression_model.fit(x_train_scaled, y_train)
```

```python
regression_model.score(x_train_scaled, y_train)
```

```python
reg_pradict=regression_model.predict(x_test_scaled)
```

```python
# Error

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test, reg_pradict))
print(mean_absolute_error(y_test, reg_pradict))
```

```python
# root mean error
print(np.sqrt(mean_squared_error(y_test, reg_pradict)))
```
